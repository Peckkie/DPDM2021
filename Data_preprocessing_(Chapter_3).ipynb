## HW12: การจัดการข้อมูลที่ซ้ำกัน ใน province_of_isolation

````python
#checking isnull หลัง ก่อน fillna
print('Is null: ', len(data_covid_cases[data_covid_cases['province_of_isolation'].isnull()]))
````

````python
#สังเกตค่า "province_of_isolation" ที่ --NaN-- แล้ว 'province_of_onset' --not NaN--
data_covid_cases[data_covid_cases['province_of_isolation'].isnull()]
````

````python
#step:1 fill missing--> 'None 'to province_of_isolation
data_covid_cases[['province_of_isolation_']] = data_covid_cases[['province_of_isolation']].fillna('None') #เติมด้วยค่า None
data_covid_cases
````

````python
#step:2 เติมค่า'province_of_onset' --not NaN-- ใน  "province_of_isolation" ที่ --NaN--
isolation = list(data_covid_cases.province_of_isolation_)
onset = list(data_covid_cases.province_of_onset)
province_of_isolation_ =[]
for i in range(len(data_covid_cases)):
    if isolation[i] == 'None':
        province_of_isolation_.append(onset[i])
    else:
        province_of_isolation_.append(isolation[i])
data_covid_cases[['province_of_isolation_']] = province_of_isolation_
#checking isnull หลัง fillna ด้วย 'province_of_onset' --not NaN--
print('Is null: ', len(data_covid_cases[data_covid_cases['province_of_isolation_'].isnull()]))
````

````python
#step:3 fillna missing ที่เหลือด้วย ค่า Mode

    #findind Mode of province_of_isolation
print((data_covid_cases[['province_of_isolation_']].dropna().mode().iloc[0,0]))#Mode
data_covid_cases[['province_of_isolation_']] = data_covid_cases[['province_of_isolation_']].fillna(data_covid_cases[['province_of_isolation_']].dropna().mode().iloc[0,0]) #เติมด้วยค่า Mode
    #checking isnull หลัง fillna ด้วย 'province_of_onset' --not NaN--
print('Is null: ', len(data_covid_cases[data_covid_cases['province_of_isolation_'].isnull()]))
````
