## 2) Create bin

````python
import numpy as np
````

````python
bin = 10 #กำหนดให้เป็น 10
#เปลี่ยนที่เดียวคือ bin
bin_r = []
for i in range(bin):
    step = int(len(my_data_sorted)/bin)
    this_data = my_data_sorted[(step*i) : (step*i)+step]
    rep = np.mean(this_data)
    bin_r.append(rep)
````

* bin1 -> 0:50            i = 0    i : i + 50
* bin2 -> 50:100         i = 1   i* 50 : (i * 50) + 50
* bin3 -> 100:150       i = 2   i* 50 : (i * 50) + 50  . . . 




````python
bin_r
````

````python
bin = 10 #กำหนดให้เป็น 10
#เปลี่ยนที่เดียวคือ bin
step = int(len(my_data_sorted)/bin)
bin_r = []
for start in range(0, len(my_data_sorted),50):
    this_data = my_data_sorted[start: start+step] #ตัดข้อมูล
    rep = np.mean(this_data) #เลือกตัวแทนของข้อมูล
    bin_r.append(rep)
````

````python
bin_r
````

## 3) Smooth

````python
#เเทนค่าข้อมูลเก่าด้วย ค่ากลาง ในแต่ละกลุ่มที่เราหามาในแต่ละ bin  ******************
bin =10 
my_data_smoothed =[]
step = 0 
this_bin = 0
for i in my_data_sorted:
    if step < 50: #ถ้า   step < 50 
        my_data_smoothed.append({'old_v':i, 'new_v':bin_r[this_bin]}) #เอาค่าข้อมูลเดิมไปใส่ไว้ old_v และ เอาค่ากลางไปใส่ไว้ new_v
        step = step+1
    else: #เมื่อวิ่งครบ 50 รอบ
        step = 0
        this_bin = this_bin+1
        my_data_smoothed.append({'old_v':i, 'new_v':bin_r[this_bin]})
````

````python
import pandas as pd
````

````python
my_df = pd.DataFrame(my_data_smoothed) #แปลง  Dict to DataFrame
my_df
````

````python
plt.scatter(range(len(my_df)),my_df['new_v']) #ลักษณะข้อมูลเเบบใหม่ที่หาจาก bin
````
