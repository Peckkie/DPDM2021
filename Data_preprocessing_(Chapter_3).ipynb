# Handling Missing Value1 (ลบค่า missing)

```python
data_covid_cases.shape
```

```python
data_covid_cases_M1 = data_covid_cases.dropna() #คำสั่งลบ null->  .dropna()
data_covid_cases_M1.shape
```

size before drop = 1000

size after drop = 10

we lost 99% of data

```python
print(f'size before drop = {len(data_covid_cases)}')
print(f'size after drop = {len(data_covid_cases_M1)}')
print(f'we lost {((len(data_covid_cases)-len(data_covid_cases_M1))/len(data_covid_cases))*100:.2f}% of data')
```

```python
((len(data_covid_cases)-len(data_covid_cases_M1))/len(data_covid_cases) )*100
```
