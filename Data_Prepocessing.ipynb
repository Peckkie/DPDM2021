```python
Population= pd.read_csv('DataFromWeb/CountriesbyPopulation.csv')
print(Population.shape)
Population.head()
```

```python
GDP= pd.read_csv('DataFromWeb/National_Debt_to_GDP_Ratio.csv')
print(GDP.shape)
GDP =GDP.rename(columns={"country": "Country"})
GDP.head()
```

# Merge table

โดยรวมที่ละตาราง เเละจะยังไม่รวมตารางที่มีข้อมูลเป็น Time series

### รวมตารางโดยเริ่มจาก ตาราง Population, MedianAge, GDP ด้วยคำสั่ง merge [PD]

```python
#รวมตารางโดยเริ่มจาก ตาราง Population, MedianAge, GDP ด้วยคำสั่ง merge [PD]

res_df = pd.merge(left=Population, right=MedianAge, how='left', on=['Country'])
print(res_df.shape[0])
res_df
```

```python
res_df = pd.merge(left=res_df, right=GDP, how='left', on=['Country'])
print(res_df.shape[0])
res_df
```

```python
res_df.isnull().sum()
```

### รวมตารางที่ 2  ตาราง Net_Change, life_expectancy, worldometer ด้วยคำสั่ง merge [PD]

```python
res_df2 = pd.merge(left=Net_Change , right=worldometer , how='left', on=['Country'])
print(res_df2.shape[0])
res_df2
```

```python
res_df2 = pd.merge(left=res_df2 , right=life_expectancy  , how='left', on=['Country'])
print(res_df2.shape[0])
res_df2
```

```python
res_df2.isnull().sum()
```

### รวม 2 ตาราง

```python
res_df = pd.merge(left=res_df, right=res_df2, how='left', on=['Country'])
print(res_df.shape[0])
res_df
```

```python
res_df.isnull().sum()
```

```python
res_df.to_csv('data_berfore_clean.csv', index=False)
```
